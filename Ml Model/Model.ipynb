```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./natural_disaster_human_mobility.csv')
df.head()
```

```python
df_bohol = df.loc[df['disaster.event'] == '12_Bohol']
df_bohol.head()
```

```python
df_bohol.describe()
```

```python
df_bohol.loc[:, ['latitude', 'longitude.anon']] = df_bohol.loc[:, ['longitude.anon', 'latitude']].values
df_bohol
```

```python
df_bohol.reset_index(inplace=True, drop=True)
df_bohol
```

```python
df_bohol['timestamp'] = df_bohol.apply(lambda row: pd.to_datetime(row['time'], format="%Y-%m-%d %X").timestamp(), axis=1)
df_bohol
```

```python
date_event_start = pd.to_datetime('2013-10-15 00:00:00', format="%Y-%m-%d %H:%M").timestamp()

df_filtered = df_bohol.loc[df_bohol['timestamp'] >= date_event_start]
df_filtered.reset_index(inplace=True, drop=True)
df_filtered.head()
```

```python
df_filtered.describe()
```

```python
print('minimum: ', pd.to_datetime(1.381795e+09, unit='s'))
print('mean: ', pd.to_datetime(1.382464e+09, unit='s'))
print('maximum: ', pd.to_datetime(1.383188e+09, unit='s'))
print()
print('25%: ', pd.to_datetime(1.382110e+09, unit='s'))
print('50%: ', pd.to_datetime(1.382450e+09, unit='s'))
print('75%: ', pd.to_datetime(1.382816e+09, unit='s'))

df_filtered.boxplot(column=['timestamp'])
```

```python
df_filtered['user.origints'] = df_filtered.apply(lambda row: min(df_filtered.loc[df_filtered['user.anon'] == row['user.anon']]['timestamp']), axis=1)
df_filtered['user.destinationts'] = df_filtered.apply(lambda row: max(df_filtered.loc[df_filtered['user.anon'] == row['user.anon']]['timestamp']), axis=1)
df_filtered
```

```python
df_user_origin = df_filtered[df_filtered['user.origints'] == df_filtered['timestamp']][['user.anon', 'latitude', 'longitude.anon']]
df_user_origin
```

```python
df_user_destination = df_filtered[df_filtered['user.destinationts'] == df_filtered['timestamp']][['user.anon', 'latitude', 'longitude.anon']]
df_user_destination
```

```python
df_user = df_user_origin.copy()
df_user = df_user.join(df_user_destination.set_index('user.anon'), on="user.anon", lsuffix="_origin", rsuffix="_dest")
df_user.columns = ['user', 'origin_lat', 'origin_long', 'dest_lat', 'dest_long']
df_user.reset_index(inplace=True, drop=True)
df_user.head()
```

```python
import matplotlib.pyplot as plt
df_user.plot(kind="scatter", x="origin_long", y="origin_lat", alpha=0.4)
plt.show()
```

```python
df_user.plot(kind="scatter", x="dest_long", y="dest_lat", alpha=0.4)
plt.show()
```

```python
def calculateDistance(lat1, long1, lat2, long2):
    R = 6371
    lat1 = lat1 * (np.pi/180)
    lat2 = lat2 * (np.pi/180)
    dLat = (lat2 - lat1) * (np.pi/180)
    dLong = (long2 - long1) * (np.pi/180)
    a = np.sin(dLat/2) * np.sin(dLat/2) + np.cos(lat1) * np.cos(lat2) * np.sin(dLong/2) * np.cos(dLong/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    d = R * c
    return d
df_user['dist (km)'] = df_user.apply(lambda row: calculateDistance(row['origin_lat'], row['origin_long'], row['dest_lat'], row['dest_long']), axis=1)
df_user.head()
```
